```python
from qiskit import QuantumCircuit
from qiskit.visualization import *
from qiskit_aer import Aer
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0, 1)

qc.draw()
```

Alice

```python
bit = '01'

if bit=='00':
  pass
elif bit=='01':
  qc.x(0)
elif bit=='10':
  qc.z(0)
elif bit=='11':
  qc.x(0)
  qc.z(0)
```

Bob

```python
# Bob decodes
qc.cx(0,1)
qc.h(0)

qc.measure(0,0)
qc.measure(1,1)
```

```python
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc,shots=1).result()
counts = result.get_counts()
print(counts)
```
